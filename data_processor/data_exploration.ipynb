```python
import pandas as pd
df = pd.read_csv(
    "detail-records/detail_record_2017_01_02_08_00_00",
    header=None
)
```

```python
print(df.head())
```

```python
df = df.dropna(axis=1, how='all')
#since last column is empty, we can drop it
```

```python
df.columns = [
    "driverID",
    "carPlateNumber",
    "Latitude",
    "Longitude",
    "Speed",
    "Direction",
    "siteName",
    "Time",
    "isRapidlySpeedup",
    "isRapidlySlowdown",
    "isNeutralSlide",
    "isNeutralSlideFinished",
    "neutralSlideTime",
    "isOverspeed",
    "isOverspeedFinished",
    "overspeedTime",
    "isFatigueDriving",
    "isHthrottleStop",
    "isOilLeak"
]
# add the column names to the dataframe
```

```python
# Check unique drivers and their car plate numbers
unique_drivers = df['driverID'].unique()
print(f"Total number of drivers: {len(unique_drivers)}")
print(f"Driver IDs: {unique_drivers}")

# Optionally check if each driver is associated with a single car plate
driver_car_mapping = df.groupby('driverID')['carPlateNumber'].unique()
print("\nDriver to Car Plate mapping:")
print(driver_car_mapping)
```
